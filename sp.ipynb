```python
from typing import List, Dict, Tuple, Set
import matplotlib.pyplot as plt
import networkx as nx
edge = Tuple[int, int]
graph = Dict[int, List]
inf = float('inf')

def Bellman_Ford(G: graph, a: List[List[int]], s: int) -> Tuple[int, List[int]]:
    d = {v: inf for v in G}
    p = {v: -1 for v in G}

    d[s] = 0

    for i in range(len(G)-1):
        for u in G:
            for v in G[u]:
                if d[v] > d[u] + a[u][v]:
                    d[v] = d[u] + a[u][v]
                    p[v] = u
    return sum(d.values()), p


```
